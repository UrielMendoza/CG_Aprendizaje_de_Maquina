Se importan la librerias necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Se Define la matriz A y el vector a para la función f1

```python
A = np.array([[10, -0.5, -5], [-0.5, 10, -11], [-5, -11, -4]])
a = np.array([5, -2, 6])
constant = 9
```

Se define la función cuadrática f1

```python
def f1(x):
    return x.T @ A @ x + a.T @ x + constant
```

Se define el gradiente de la función cuadrática f1

```python
def grad_f1(x):
    return 2 * A @ x + a
```

Se crear una cuadrícula de puntos (solo en 2D para visualización) para poder visualizar los vectores.

```python
x1_values = np.linspace(-5, 5, 100)
x2_values = np.linspace(-5, 5, 100)
X1, X2 = np.meshgrid(x1_values, x2_values)
F = np.zeros_like(X1)

for i in range(len(x1_values)):
    for j in range(len(x2_values)):
        x_vec = np.array([x1_values[i], x2_values[j], 0])  # f1 en 2D
        F[i, j] = f1(x_vec)
```

Se crea la visualización de la función en 2D

```python
plt.figure(figsize=(10, 8))
contour = plt.contourf(X1, X2, F, cmap='viridis', levels=50, alpha=0.5)
plt.colorbar(contour)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title(r'Función $f_1(x_1, x_2, 0)$')
```

Se añade la visualización de los gradientes en 2D (invertidos para apuntar al mínimo)

```python
plt.figure(figsize=(10, 8))
contour = plt.contourf(X1, X2, F, cmap='viridis', levels=50, alpha=0.5)
plt.colorbar(contour)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title(r'Función $f_1(x_1, x_2, 0)$')
for i in range(0, len(x1_values), 10):
    for j in range(0, len(x2_values), 10):
        x_vec = np.array([x1_values[i], x2_values[j], 0])  # f1 en 2D
        grad = -grad_f1(x_vec)  # Invertir gradiente
        plt.quiver(x1_values[i], x2_values[j], grad[0], grad[1], angles='xy', scale_units='xy', scale=100, color='red', width=0.003)

plt.show()
```

Se realiza la visualización de la función en 3D

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, F, cmap='viridis', edgecolor='none', alpha=0.5)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$f_1(x_1, x_2, 0)$')
ax.set_title(r'Función $f_1(x_1, x_2, 0)$ en 3D')
```

Se añade la visualización de los gradientes en 3D (invertidos para apuntar al mínimo)

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, F, cmap='viridis', edgecolor='none', alpha=0.5)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$f_1(x_1, x_2, 0)$')
ax.set_title(r'Función $f_1(x_1, x_2, 0)$ en 3D')

for i in range(0, len(x1_values), 10):
    for j in range(0, len(x2_values), 10):
        x_vec = np.array([x1_values[i], x2_values[j], 0])  # f1 en 2D
        grad = -grad_f1(x_vec)  # Invertir gradiente
        ax.quiver(x1_values[i], x2_values[j], F[i, j], grad[0], grad[1], 0, color='red', length=0.5, normalize=True)

plt.show()
```
