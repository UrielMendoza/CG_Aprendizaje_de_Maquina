Se importan la librerias necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
```

Se Definie la matriz Sigma como una matriz simétrica y semidefinida positiva

```python
sigma = np.array([[2, 0], [0, 1]])
```

Se define una funcion parra para calcular la norma $\|x\|_\Sigma$

```python
def norm_Sigma(x, Sigma):
    return np.sqrt(np.dot(x.T, np.dot(Sigma, x)))
```

Se crea una serie de vectores x para calcular la norma


```python
vectors = [np.array([1, 1]), np.array([2, 2]),np.array([-1, -1]), np.array([-2, -2]) ]
```

Se crea la grafica con los vectores y la norma calculada

Norma $\|x\|_\Sigma = \sqrt{x^T \Sigma x}$ con $\Sigma = \begin{bmatrix} 2 & 0 \\ 0 & 1 \end{bmatrix}$

```python
plt.figure(figsize=(8, 8))
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

for vector in vectors:
    norm = norm_Sigma(vector, sigma)
    plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='blue')
    plt.text(vector[0], vector[1], f'{norm:.2f}', fontsize=12, ha='right')

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
#plt.title(r'Norma $\|x\|_\Sigma = \sqrt{x^T \Sigma x}$ con $\Sigma = \begin{bmatrix} 2 & 0 \\ 0 & 1 \end{bmatrix}$')
plt.show()
```
