Se importan la librerias necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
```

Se define la función para graficar la transformación afín definida por una rotación R, escalado S, y cizallamiento H.

```python
def plot_transformation(theta, s1, s2, h, tx, ty):
    '''
    Función para graficar la transformación afín definida por una rotación R, escalado S, y cizallamiento H.
    Parámetros:
    theta -- Ángulo de rotación en radianes.
    s1 -- Escalado en dirección x.
    s2 -- Escalado en dirección y.
    h -- Cizallamiento en dirección x.
    tx -- Traslación en dirección x.
    ty -- Traslación en dirección y.
    '''

    plt.style.use('seaborn-dark-palette')

    # Se define la matriz de rotación R, escalado S, y cizallamiento H
    R = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
    S = np.array([[s1, 0], [0, s2]])
    H = np.array([[1, h], [0, 1]])

    # Se combinan las transformaciones en una matriz A
    A = R @ S @ H

    # Se define el vector de traslación b
    b = np.array([tx, ty])

    # Se definen los puntos originales del cuadrado
    square = np.array([[0, 1, 1, 0, 0], [0, 0, 1, 1, 0]])

    # Se aplica la transformación afin
    transformed_square = A @ square + b[:, np.newaxis]

    # Se grafican los dos cuadrados
    plt.figure(figsize=(6, 6))
    plt.plot(square[0, :], square[1, :], color='black', linestyle='-', linewidth=1, label='Original')
    plt.plot(transformed_square[0, :], transformed_square[1, :], color='black', linestyle='-', linewidth=1, label='Transformado')
    plt.axis('equal')
    plt.title('Transformación Afin, Rotación R, Escalado S, y Cizallamiento H')
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.legend()
    plt.grid(True)
    plt.show()
```

Se prueba la función con los parametros de rotación, escalado, cizallamiento y traslación lo suficientemente para notar cambion en la figura.

```python
# Parametros de trasformación
# Rotación R
theta = np.pi/4
# Escalado S
s1 = 2
s2 = 1
# Cizallamiento H
h = 2
# Traslación
tx = 1
ty = 2

plot_transformation(theta, s1, s2, h, tx, ty)
```
